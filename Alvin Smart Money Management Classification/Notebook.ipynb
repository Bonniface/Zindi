## Convert Boolean to Binary

```python
# Based on our data set, this function converts boolean to binary entries
def create_binary_cols(content):
  if content == False:
    content = 0
  elif content == True:
    content = 1
  elif content == 'N':
    content = 0
  elif content == 'Y':
    content = 1
  elif content == 'Male':
    content = 0
  elif content == 'Female':
    content = 1
  return content
```

```python
# Gender column convert:
train['USER_GENDER'] = train['USER_GENDER'].apply(create_binary_cols)
test['USER_GENDER'] = test['USER_GENDER'].apply(create_binary_cols)

# Is_purchase_paid_via_mpesa_send_money column convert:
train['IS_PURCHASE_PAID_VIA_MPESA_SEND_MONEY'] = train['IS_PURCHASE_PAID_VIA_MPESA_SEND_MONEY'].apply(create_binary_cols)
test['IS_PURCHASE_PAID_VIA_MPESA_SEND_MONEY'] = test['IS_PURCHASE_PAID_VIA_MPESA_SEND_MONEY'].apply(create_binary_cols)
```

```python
encoding_dict = dict()
for col in train:
    if train[col].dtype == 'object':
        #get unique values in the column to encode
        values = train[col].value_counts().index.tolist()

        # create a dictionary of values and corresponding number {value, number}
        dict_values = {value: count for value, count in zip(values, range(1,len(values)+1))}

        # save the values to encode in the dictionary
        encoding_dict[col] = dict_values

        # replace the values with the corresponding number from the dictionary
        train[col] = train[col].map(lambda x: dict_values.get(x)) - 1
        if col != 'MERCHANT_CATEGORIZED_AS':
            test[col] = test[col].map(lambda x: dict_values.get(x)) - 1 

dict_values
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from catboost import Pool
train_pool = Pool(X_train, y_train)
val_pool = Pool(X_val, y_val)
```

```python
from catboost import CatBoostRegressor
catboost_base = CatBoostClassifier(
    # loss_function='MultiLogloss',
    # eval_metric='HammingLoss',
    iterations=500,
    depth=10,
    loss_function='MultiClass',
    # eval_metric='Logloss',
    random_seed = 42,
    bagging_temperature=0.2,
    od_type='Iter',
    od_wait=20
    # cat_features= Cat_Features
    # class_names=['Bills & Fees',
    #                 'Groceries',
    #                 'Data & WiFi',
    #                 'Miscellaneous',
    #                 'Family & Friends',
    #                 'Going out'
    #                 'Transport & Fuel',
    #                 'Shopping',
    #                 'Emergency fund',
    #                 'Health',
    #                 'Loan Repayment',
    #                 'Rent / Mortgage',
    #                 'Education'
    #             ]
)
```

```python
catboost_base.fit(train_pool,eval_set=val_pool,plot=True,metric_period=10,verbose=50)
```

```python
from catboost.utils import eval_metric
val_pred =  catboost_base.predict(X_val)
# Val_pred =  catboost_base.predict_proba(X_val)
Accuracy= eval_metric(y_val,val_pred,'Accuracy')[0]
print(f'Score: {Accuracy}')
```

```python
# Fit model
model = catboost_base
model.fit(train_pool)

# Get predicted classes
preds_class = model.predict(val_pool)

# Get predicted probabilities for each class
preds_proba = model.predict_proba(val_pool)

# Get predicted RawFormulaVal
preds_raw = model.predict(val_pool, 
                          prediction_type='RawFormulaVal')
model.fit(train_pool,
          use_best_model=True,
          eval_set=val_pool,plot=True)
print("Count of trees in model = {}".format(model.tree_count_))

```
