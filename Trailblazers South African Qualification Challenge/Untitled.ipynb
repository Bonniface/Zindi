```python
Train = Train.drop(['target_min', 'target_max',
       'target_variance', 'target_count'],axis=1)

Train.shape
```

```python
X = Train.drop(['target','Date', 'Place_ID'],axis=1)
y = Train.target
Test = Test.drop(['Date', 'Place_ID'],axis=1)
X.shape, y.shape ,Test.shape
```

```python
X.describe().transpose()
```

```python
from sklearn.impute import SimpleImputer


for col in X.select_dtypes('number'):
    cat_imputer = SimpleImputer(strategy='mean')
    cat_imputer.fit(X[[col]])
    X.loc[:, col] = cat_imputer.transform(X[[col]])
    Test.loc[:, col] = cat_imputer.transform(Test[[col]])
```
